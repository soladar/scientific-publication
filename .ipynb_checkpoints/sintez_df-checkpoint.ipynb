```python
import numpy as np
import pandas as pd
```

```python
data = {}
num_rows = 374

#Параметры для каждого стобца
distributions = [
    {'name': 'normal', 'mean': 5, 'std': 2}, # X_1
    {'name': 'uniform', 'low': 0, 'high': 10}, # X_2
    {'name': 'uniform', 'low': 16, 'high': 50}, # X_3
    {'name': 'uniform', 'low': 0, 'high': 2} # X_4
]

for i, dist in enumerate(distributions):
    distribution_name = dist['name']

    if distribution_name == 'normal':
        mean = dist['mean']
        std = dist['std']
        column_data = np.random.normal(loc=mean, scale=std, size=num_rows)
    elif distribution_name == 'uniform':
        low = dist['low']
        high = dist['high']
        column_data = np.random.uniform(low=low, high=high, size=num_rows).round()
    data[f'X_{i + 1}'] = column_data
    
df = pd.DataFrame(data)

df.head()
```

```python
#Нормальное распределение с одним пиковым значением

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df['Peak_Column'] = df['X_1'] + np.abs(np.random.normal(loc=0, scale=3, size=num_rows))

sns.histplot(data=df['Peak_Column'], kde=True)
plt.title('Peak column')
plt.show()
```

```python
#Функц. для создания разных распределений по категориям

def generate_feature(category):
    if category == 0:
        return np.random.normal(loc=5, scale=2)
    elif category == 1:
        return np.random.uniform(low=0, high=10)
    elif category == 2:
        return np.random.gamma(shape=2, scale=2)
```

```python
df['Correlated_Column_X1_1'] = df['X_1'] + np.random.normal(loc=0, scale=1, size=num_rows) #Столбец с корреляцией по столбцу X1
df['Correlated_Column_X1_2'] = df['X_1'] + np.random.normal(loc=0, scale=1, size=num_rows) #Столбец с корреляцией по столбцу X1
df['Binnary_to_X3'] = df['X_3'].apply(lambda x: 1 if x > 40 else 0) #Столбец с бинарными признаками, если число в X_3 > 40, то 1
df['Numbers_to_X4'] = df['X_4'].apply(generate_feature) #Разное распределение по категориям X3
df['Multiply_X1_X2'] = df['X_1'] * df['X_2'] #Столбец значения которого ранвны X_1 умножить X_2
df['Binary_Mode_X_3'] = df['X_3'].apply(lambda x: 1 if x == int(df['X_3'].mode()) else 0) #Бинарный признак,                                                                                        #если значение по X_3 равно моде, то 1
df['X3/(Peak_Column)'] = df['X_3']/df['Peak_Column'] #Столбец где X_3 разделили Peak_Column
df['(Correlated_Column_X1_1) - X_1'] = df['Correlated_Column_X1_1'] - df['X_1'] # Разница между Correlated_Column_X1_1 и X_1

df
```

```python
#Распределение в столбце Numbers_to_X4 по категориям в X_4

for category in df['X_4'].unique():
    table = df[['Numbers_to_X4', 'X_4']].query(f'X_4 == {category}')
    sns.histplot(data=table['Numbers_to_X4'], kde=True)
    plt.title(f'распределение {int(category)} категории')
    plt.show()
```

```python
#Тепловая карта корреляции

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Гистограммы по столбцам

for column in df.columns:
    sns.histplot(data=df[column], kde=True)
    plt.title(column)
    plt.show()
```

```python
df.to_csv('sint_dataset.csv', index=False)
```
